```python
def imprimir(x):
    if x > 0:
        imprimir(x-1)
        print(x)

imprimir(3)
```

```python
#Suma de una lista sin recursividad
def suma(lista):
    suma = 0
    for i in range(0,len(lista)):
        suma = suma + lista[i]
    return suma

print(suma([6,3,4,2,10]))
```

```python
#Version recursiva

def suma(lista):
    if len(lista) == 1:
        return lista[0]
    else:
        return lista[0] + suma(lista[1:])

print(suma([6,3,4,2,10]))
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(10))
```
